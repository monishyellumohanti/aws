```python
plt.figure(1)
plt.subplot(2,2,1)
df["engine-type"].value_counts().plot(figsize=(10,8),kind="bar",color="red")
plt.title("Engine Type Frequence Diagram")
plt.ylabel("No. of Engine type")
plt.xlabel("Engine Type")

plt.subplot(2,2,2)
df["drive-wheels"].value_counts().plot(figsize=(10,8),kind="bar",color="green")
plt.title("Drive wheels Type Frequence Diagram")
plt.ylabel("No. of Drive wheels type")
plt.xlabel("Drive Wheels Type")

plt.subplot(2,2,3)
df["fuel-type"].value_counts().plot(figsize=(10,8),kind="bar",color="orange")
plt.title("fuel Type Frequence Diagram")
plt.ylabel("No. of Fuel type")
plt.xlabel("Fuel Type")

plt.subplot(2,2,4)
df["body-style"].value_counts().plot(figsize=(10,8),kind="bar",color="purple")
plt.title("Body Style Frequence Diagram")
plt.ylabel("No. of Body Style type")
plt.xlabel("Body Style Type")

plt.tight_layout()
plt.show()
```

```python
df["body-style"].value_counts().plot(figsize=(6,6),kind="pie")
```

```python
explode=[0,0,0.1,0,0]
df["body-style"].value_counts().plot(figsize=(6,6),kind="pie",autopct="%.2f%%",explode=explode)
plt.title("body-style pie charrt")
plt.xlabel("type of vehicle")
plt.ylabel("body style type")
```
